# Netflix Movies
## Are movies becoming shorter?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv('netflix_data.csv')
netflix_df
```

```python
netflix_subset = netflix_df[netflix_df['type'] != 'TV Show']
netflix_subset
```

```python
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
netflix_movies
```

```python
short_movies = netflix_movies[netflix_movies['duration'] < 60]
short_movies
```

#### 'Documentaries' and 'Children' movies make up the majority of short movies.

```python
short_movies.groupby('genre')['genre'].value_counts().sort_values(ascending=False)
short_movies.groupby('genre')['genre'].value_counts().sort_values(ascending=False).plot(kind='barh')
```

```python
short_movies.loc[~short_movies['genre'].isin(['Documentaries', 'Children', 'Stand-Up']), 'genre'] = 'Other'
short_movies
```
