# Netflix Movies
## Are movies becoming shorter?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv('netflix_data.csv')
netflix_df
```

```python
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
netflix_subset
```

```python
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
netflix_movies
```

```python
short_movies = netflix_movies[netflix_movies['duration'] < 60]
short_movies
```

#### 'Documentaries' and 'Children' movies make up the majority of short movies.

```python
netflix_movies.groupby('genre')['genre'].value_counts().sort_values(ascending=False)
netflix_movies.groupby('genre')['genre'].value_counts().sort_values(ascending=False).plot(kind='barh')
```

```python
# df["C"][df["C"]=="foo3"] = "foo33"
# above causes 'SettingWithCopyWarning'
# df.loc[df["C"]=="foo3", "C"] = "foo333"

netflix_movies.loc[~netflix_movies['genre'].isin(["Children", "Documentaries", "Stand-Up"]), 'genre'] = "Other"
```

```python
netflix_movies.groupby('genre')['genre'].value_counts().sort_values(ascending=False).plot(kind='barh')
```

```python
colors = []

# Iterate over rows of netflix_movies
for label, row in netflix_movies.iterrows() :
    if row["genre"] == "Children" :
        colors.append("red")
    elif row["genre"] == "Documentaries" :
        colors.append("blue")
    elif row["genre"] == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
```

```python
fig = plt.figure(figsize=(18, 10))

plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)

plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

plt.show()
```
